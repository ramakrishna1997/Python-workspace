```python
from sklearn.model_selection import train_test_split
```

```python
tip.columns
```

```python
x=tip[['total_bill','size']]
```

```python
x
```

```python
y=tip[['tip']]
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
plt.plot(X_train)
```

```python
data=pd.DataFrame(lm.coef_.T,x.columns,columns=['coef'])
```

```python
data
```

```python
data.plot()
```

```python
predicts=lm.predict(X_test)
```

```python
plt.plot(y_test,predicts)
```

```python
sns.distplot(y_test-predicts)
```

```python
from sklearn import metrics
```

```python
MAE=metrics.mean_absolute_error(y_test,predicts)
```

```python
MSE=metrics.mean_squared_error(y_test,predicts)
```

```python
RMSE=np.sqrt(metrics.mean_squared_error(y_test,predicts))
```

```python
print(MAE)
print(MSE)
print(RMSE)
```
