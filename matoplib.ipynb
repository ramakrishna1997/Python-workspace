# pandas and visualize data

```python
a1=np.random.randn(20,5)
b1=np.random.randn(20,5)
c1=[a1,b1]
```

```python
dfa=pd.DataFrame(a1)
dfb=pd.DataFrame(b1)
```

```python
#dfa
dfb
```

```python
dfc=pd.concat([dfa,dfb])
```

```python
dfc.shape
```

```python
dfc.columns=['A','B','C','D','E']
```

```python
dfc
```

```python
dfc
```

```python
dfc.index.name="Slno"
```

```python
plt.plot(dfc['A'],dfc['D'],color='y',ls='--',lw=3,marker='d')
plt.plot(dfc['C'],dfc['B'],color='pink',ls='dotted',lw=5,marker='o')
plt.plot(dfc['D'],dfc['E'],color='r',ls='dashed',lw=6,marker='^')
plt.title("info")
plt.legend(["A","B","C","D"],loc='lower left')
plt.grid()
```
